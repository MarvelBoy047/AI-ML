```python
years = range(2000,2008)
apples = [235,255,666,363,456,452,169,251]
oranges=[365,258,554,655,255,665,259,263]
plt.plot(years,apples)
plt.plot(years,oranges)
plt.xlabel('year')
plt.xlabel('yield')
# plt.plot(year,apples,marker = 'o')
# plt.plot(year,oranges,marker = 'x')
plt.plot(years,apples,'s-g')
plt.plot(years,oranges,'o--r')
plt.title('Crop yield in kanto')
plt.legend(['Apples','Oranges'])


```
